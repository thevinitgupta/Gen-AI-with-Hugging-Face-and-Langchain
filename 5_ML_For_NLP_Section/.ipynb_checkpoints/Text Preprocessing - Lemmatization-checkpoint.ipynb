## Wordnet Lemmatizer
Lemmatization technique is like stemming. The output we will get after lemmatization is called ‘lemma’, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the [wordnet](https://wordnet.princeton.edu/) corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example −


```python
## Used for Q&A, Chatbots, Text Summarization
import nltk
nltk.download('wordnet')

lemmatizer = nltk.WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going") #default Part of Speech(POS) = n (Noun)
```

```python
lemmatizer.lemmatize("going",pos='v') # Part of Speech(POS) = v (Verb)
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized"]

for word in words:
    print(word + " ----> " + lemmatizer.lemmatize(word))
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized"]

for word in words:
    print(word + " ----> " + lemmatizer.lemmatize(word, pos='v'))
```
